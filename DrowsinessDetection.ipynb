```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
# Load YOLOv11 Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
# Make Detections with Images
img = 'D:\Jupyter\DrowsinessDetection\TalkingOnPhone.png'
results = model(img)
results.print()

# Display Results
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Mirror the frame
    mirrored_frame = cv2.flip(frame, 1)

    # Make detections on the mirrored frame
    results = model(mirrored_frame)
    
    # Show the results
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    # Exit when 'q' is pressed
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import uuid #Unique identifier
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images') #/data/images
labels = ['awake', 'drwosy']
number_imgs = 20 #Number of images we want to collect
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        
        # Writes out image to file 
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO

# Load the YOLOv8 custom model
model = YOLO(r"D:\Jupyter\DrowsinessDetection\ultralytics\runs\detect\train4\weights\last.pt")

# Path to the image for inference
image_path = r"data/images/drwosy.75e9732f-b6fc-11ef-9c9c-e02e0b7b90e3.jpg"

# Perform inference
results = model.predict(source=image_path, conf=0.25)

# Iterate over the results to display or save
for result in results:
    result.plot()  # Annotates and displays the image

# Print detection results (bounding boxes, confidence scores, etc.)
for result in results:
    print(result.boxes)  # Bounding box information
```
